```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

website_traffic = pd.read_csv("website_traffic.csv")
website_traffic
```

```python
website_traffic.info()
```

```python
new_website_traffic = website_traffic.drop('Traffic Source', axis=1)
```

```python
new_website_traffic
```

```python
k = 6

correlation = new_website_traffic.corr(numeric_only=True)
cols = correlation.nlargest(k, "Conversion Rate")['Conversion Rate'].index

print(cols)
```

```python
cm = np.corrcoef(new_website_traffic[cols].values.T)

fig, ax = plt.subplots(figsize=(14,12))
sns.heatmap(cm, vmax=.8, linewidth=0.01, square = True, annot = True, cmap='viridis', xticklabels=cols.values, annot_kws = {'size': 12}, yticklabels=cols.values)
```

```python
# feature selection
new_website_traffic_features = ["Page Views", "Session Duration", "Bounce Rate", "Time on Page"]

x = new_website_traffic[new_website_traffic_features]
x
```

```python
y = website_traffic['Conversion Rate']
y
# print(y.head(114))
```

```python
from sklearn.tree import DecisionTreeRegressor

website_traffic_model = DecisionTreeRegressor(random_state = 1)
website_traffic_model.fit(x, y)
```

```python
from sklearn.metrics import mean_absolute_error

print("Making predictions based on the following feature selected dataset: ")
# print(x.head())
print(x)

# prediction = melb_model.predict(x.head())
prediction = website_traffic_model.predict(x)
print("\nPredicted values of the Conversion Rate: ")
print(prediction)

error = mean_absolute_error(y, prediction)
print("\nThe Conversion off by", round(error), "%")
```

```python
from sklearn.model_selection import train_test_split

train_x, val_x, train_y, val_y = train_test_split(x, y, random_state = 1)

website_traffic_model = DecisionTreeRegressor(random_state = 1)
website_traffic_model.fit(train_x, train_y)
predictions = website_traffic_model.predict(val_x)

result = mean_absolute_error(val_y, predictions)

print(result)

print(f"\nIf returns a value of {result}, it implies that, on average, the model's predictions are off by {result} units (in the context of the target variable) from the actual values.")
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 4, random_state = 40)
kmeans.fit(new_website_traffic)
new_website_traffic['clusters'] = kmeans.predict(new_website_traffic)
new_website_traffic
```

```python
cluster_means = new_website_traffic.groupby('clusters').mean()
cluster_means
```

```python
# Assuming 'new_website_traffic' is a DataFrame containing the columns you need
# Replace the following line with your actual DataFrame
# new_website_traffic = pd.read_csv('your_data.csv')

# Columns to plot
columns_to_plot = ['Page Views', 'Session Duration', 'Bounce Rate', 
                   'Time on Page', 'Previous Visits', 'Conversion Rate']

# Apply KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=42)
new_website_traffic['cluster'] = kmeans.fit_predict(new_website_traffic)

# Group by clusters and calculate the mean of each metric
cluster_means = new_website_traffic.groupby('cluster').mean()

# Create subplots for each metric
fig, axes = plt.subplots(1, len(columns_to_plot), figsize=(15, 5), sharex=True)

# Plot each metric in a separate subplot
for ax, column in zip(axes, columns_to_plot):
    ax.plot(cluster_means.index, cluster_means[column], '-o')
    ax.set_title(column)
    ax.set_xlabel('Cluster')

axes[0].set_ylabel('Mean Value')

plt.tight_layout()
plt.show()
```

```python
# Analyze clusters
for cluster_id in cluster_means.index:
    print(f"Cluster {cluster_id}:")
    for column in columns_to_plot:
        mean_value = cluster_means.loc[cluster_id, column]
        print(f"  {column}: {mean_value:.2f}")

    # Example of identifying patterns
    if cluster_means.loc[cluster_id, 'Conversion Rate'] > cluster_means['Conversion Rate'].mean():
        print(f"  - This cluster has a higher conversion rate than average.")
    if cluster_means.loc[cluster_id, 'Bounce Rate'] < cluster_means['Bounce Rate'].mean():
        print(f"  - This cluster has a lower bounce rate than average.")
    if cluster_means.loc[cluster_id, 'Page Views'] > cluster_means['Page Views'].mean():
        print(f"  - This cluster has a higher number of page views than average.")
    if cluster_means.loc[cluster_id, 'Session Duration'] > cluster_means['Session Duration'].mean():
        print(f"  - This cluster tends to have longer session durations.")
    print()
```
